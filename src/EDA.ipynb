```python
def plot_violins_with_outliers(df):
    """
    Generates violin plots for numeric columns in a DataFrame, highlighting outliers, 
    with a maximum of 8 plots per row.

    Args:
        df (pd.DataFrame): The input DataFrame.
    """

    ncols = len(df.columns)
    nrows = (ncols - 1) // 8 + 1  

    fig, axes = plt.subplots(nrows=nrows, ncols=8, figsize=(20, 5 * nrows))
    axes = axes.flatten() 

    for i, col in enumerate(df.columns):
        if pd.api.types.is_numeric_dtype(df[col]):
            data = df[col]
            ax = axes[i]  

            # Calculate IQR for outlier detection
            q1 = data.quantile(0.25)
            q3 = data.quantile(0.75)
            iqr = q3 - q1

            # Identify outliers
            outliers = data[~((data >= q1 - 1.5*iqr) & (data <= q3 + 1.5*iqr))]

            # Create the violin plot
            sns.violinplot(data=data, ax=ax)

            # Plot outliers as individual points
            sns.stripplot(data=outliers, color='red', jitter=0.2, size=3, ax=ax)

            # Labels and titles
            ax.set_title(col)
            ax.set_xlabel("")
            ax.set_ylabel("Value")

    # Remove unused subplots
    for ax in axes[ncols:]:
        fig.delaxes(ax)

    plt.tight_layout()
    plt.show()
```

```python
plot_violins_with_outliers(data_1)
```

```python
plot_violins_with_outliers(data_2)
```
